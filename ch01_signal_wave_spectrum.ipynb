# Signals
- 定义信号，更加类似信号生成器

instantiate cosine ans sine signals

```python
cos_sig = thinkdsp.CosSignal(freq=440,amp=1.0,offset=0)
sin_sig = thinkdsp.SinSignal(freq=880,amp=0.5,offset=0)
```

```python
cos_sig.plot()
```

```python
mix = sin_sig + cos_sig
mix.plot()
```

# Waves
- after evaluate a signal at a sequence of times, the result is a Wave.

```python
wave = mix.make_wave(duration=0.5, start=0, framerate=11025)
wave
```

```python
wave.plot()
```

```python
from IPython.display import Audio
audio = Audio(data = wave.ys,rate=wave.framerate)
audio
```

```python
# 播放wave
wave.make_audio()
```

```python
print('Number of samples', len(wave.ys))
print('Timestep in ms', 1 / wave.framerate * 1000)
```

```python
period = mix.period
segment = wave.segment(start=0, duration=period*3)
period
```

```python
segment.plot()
thinkplot.config(xlabel='Time (s)')
```

# Spectrums

- the WAVE's spectrums

```python
spectrum = segment.make_spectrum()
```

```python
spectrum.plot()
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
spectrum.low_pass(500)
spectrum.plot()
```

convert spectrum back to wave

```python
filtered = spectrum.make_wave()
```
